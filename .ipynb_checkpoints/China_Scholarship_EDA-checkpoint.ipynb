```python
#Analysis of the columns with null values
#Tuition_Covered Columm

fig = plt.figure(figsize = (10,6))
df['Tuition_Covered'].plot(kind='hist', grid = True)
plt.title('Histogram of Tuition_Covered', fontsize = 14)
```

```python
#Original_Tuition_Fee Columm

fig = plt.figure(figsize = (10,6))
df['Original_Tuition_Fee'].plot(kind='hist', grid = True)
plt.title('Histogram of Original_Tuition_Fee', fontsize = 14)
```

```python
#Replacing missing values in columns with mean 
median_1 = df['Tuition_Covered'].median()
median_2 = df['Original_Tuition_Fee'].median()

df['Tuition_Covered'].replace(np.nan, median_1, inplace = True)
df['Original_Tuition_Fee'].replace(np.nan, median_2, inplace = True)
```

```python
df.isnull().sum()
```

The columns with null values were filled with the median

```python
plt.subplot(221)
df['Language'].value_counts().plot(kind = 'bar', title = 'Language', figsize=(12, 8))
plt.xticks(rotation = 360)
plt.subplot(222)
df['Start_Month'].value_counts().plot(kind = 'bar', title = 'Start_Month')
plt.xticks(rotation = 360)
plt.subplot(223)
df['Accomodation_Duration'].value_counts().plot(kind = 'bar', title = 'Accomodation_Duration')
plt.xticks(rotation = 360)
plt.subplot(224)
df['Expense_Duration'].value_counts().plot(kind = 'bar', title = 'Expense_Duration')
plt.xticks(rotation = 360)

plt.show()
```

```python
df['Accomodation_Covered?'].value_counts().plot(kind = 'bar', title = 'Accomodation_Covered?')
plt.xticks([0, 1], ['No', 'Yes'], rotation = 360)

plt.show()
```

```python
df['Living_Expense_Covered?'].value_counts().plot(kind = 'bar', title = 'Living_Expense_Covered?')
plt.xticks([0, 1], ['No', 'Yes'], rotation = 360)

plt.show()
```

CONCLUSION

Zhejiang Normal University has the highest frequency.

Most Universities give scholarships at the Master's level followed by the Bachelor's level.

The most common language is Chineese.

The start month of the scholarship is usually September while the Accomodation_Duration and Expense_Duration is usually monthly.

About 2500 Scholarships do not cover for accomodation(Accomodation_Covered?) and living expense(Living_Expense_Covered?)

Computer Science and Technology is the highest major for scholarships.
